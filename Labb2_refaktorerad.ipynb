```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import openpyxl as op
import numpy as np
```

```python
sheets = pd.ExcelFile("data/riket2023_åk9_np.xlsx").sheet_names
print(sheets)  
```

```python
df_english = pd.read_excel("data/riket2023_åk9_np.xlsx", sheet_name = "Engelska", header = 8) #hämtar sheet Engelska
df_math = pd.read_excel("data/riket2023_åk9_np.xlsx", sheet_name = "Matematik", header = 8) #hämtar sheet Matematik
df_swedish = pd.read_excel("data/riket2023_åk9_np.xlsx", sheet_name = "Svenska", header = 8)  #hämtar sheet Svenska
df_SAS = pd.read_excel("data/riket2023_åk9_np.xlsx", sheet_name = "Svenska som andraspråk", header = 8)  #hämtar sheet Svenska

```

```python
df_english.columns = [
    "Plats", "Huvudman", "Totalt (A-F)", "Flickor (A-F)", "Pojkar (A-F)", 
    "Totalt (A-E)", "Flickor (A-E)", "Pojkar (A-E)", 
    "Totalt (poäng)", "Flickor (poäng)", "Pojkar (poäng)"
]
df_math.columns = [
    "Plats", "Huvudman", "Totalt (A-F)", "Flickor (A-F)", "Pojkar (A-F)", 
    "Totalt (A-E)", "Flickor (A-E)", "Pojkar (A-E)", 
    "Totalt (poäng)", "Flickor (poäng)", "Pojkar (poäng)"
]
df_swedish.columns = [
    "Plats", "Huvudman", "Totalt (A-F)", "Flickor (A-F)", "Pojkar (A-F)", 
    "Totalt (A-E)", "Flickor (A-E)", "Pojkar (A-E)", 
    "Totalt (poäng)", "Flickor (poäng)", "Pojkar (poäng)"
]
df_SAS.columns = [
    "Plats", "Huvudman", "Totalt (A-F)", "Flickor (A-F)", "Pojkar (A-F)", 
    "Totalt (A-E)", "Flickor (A-E)", "Pojkar (A-E)", 
    "Totalt (poäng)", "Flickor (poäng)", "Pojkar (poäng)"
]

```

```python
print(df_english.head())
print("************")
print(df_math.head())
print("************")
print(df_swedish.head())
print("************")
print(df_SAS.head())
```

```python
df_SAS = df_SAS.replace("..", np.nan) #Ersätter .. med null i dataframe df_SAS
df_SAS =df_SAS.dropna() 
```

```python
high_score_english = df_english.groupby("Huvudman")["Totalt (poäng)"].median().sort_values(ascending=False) #Skriver ut högsta poäng baserat på huvudman
high_score_df_english = high_score_english.reset_index()

high_score_math = df_math.groupby("Huvudman")["Totalt (poäng)"].median().sort_values(ascending=False) #Skriver ut högsta poäng baserat på huvudman
high_score_df_math = high_score_math.reset_index()

high_score_swedish = df_swedish.groupby("Huvudman")["Totalt (poäng)"].median().sort_values(ascending=False) #Skriver ut högsta poäng baserat på huvudman
high_score_df_swedish = high_score_swedish.reset_index()


high_score_SAS = df_SAS.groupby("Huvudman")["Totalt (poäng)"].median().sort_values(ascending=False) #Skriver ut högsta poäng baserat på huvudman
high_score_df_SAS = high_score_SAS.reset_index()

```

```python
sns.barplot(data = high_score_df_english, x = "Huvudman", y = "Totalt (poäng)"),  #Skriver ut högsta poäng som ett diagram
```

```python
sns.barplot(data = high_score_df_math, x = "Huvudman", y = "Totalt (poäng)"), #Skriver ut högsta poäng som ett diagram
```

```python
sns.barplot(data = high_score_df_swedish, x = "Huvudman", y = "Totalt (poäng)"), #Skriver ut högsta poäng som ett diagram
```

```python
sns.barplot(data = high_score_df_SAS, x = "Huvudman", y = "Totalt (poäng)") #Skriver ut högsta poäng som ett diagram
```
